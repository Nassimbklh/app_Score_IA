```python
import pandas as pd
```

```python
# Charger les fichiers CSV
file_matches = 'matches_transformed.csv'
file_predictions = 'predicted_match_results_with_bet.csv'
file_odds = 'betting_odds_final.csv'
```

```python
# Charger les donn√©es dans des DataFrames
df_matches = pd.read_csv(file_matches)
df_predictions = pd.read_csv(file_predictions)
df_odds = pd.read_csv(file_odds)
```

```python
df_merged = pd.merge(df_odds, df_matches, left_on=['home_team', 'away_team'], right_on=['HomeTeam', 'AwayTeam'])
```

```python
df_merged
```

```python
# Fusionner les deux DataFrames
merged_df = pd.merge(df_merged, df_predictions, left_on=['home_team', 'away_team'], right_on=['HomeTeam', 'AwayTeam'])
```

```python
merged_df.drop(columns=['HomeTeam_x', 'HomeTeam_y', 'AwayTeam_x', 'AwayTeam_y'])
```

```python
# Fonction pour mettre √† jour le capital en fonction de la pr√©diction
def pari(row, mise_constante):
    if row['Prediction'] == row['Result']:
        if row['Prediction'] == 0:
            return (mise_constante * row['home_dont_win_odds']) - mise_constante
        elif row['Prediction'] == 1:
            return (mise_constante * row['home_win_odds']) - mise_constante
    else:
        return -mise_constante
```

```python
# Compter les NaN s√©par√©ment dans Prediction
nan_prediction_count = df_predictions['Prediction'].isna().sum()
non_nan_prediction_count = len(df_predictions) - nan_prediction_count
```

```python
# Supprimer les lignes avec NaN dans Prediction
merged_df = merged_df.dropna(subset=['Prediction'])
```

```python
# Initialiser la variable
mise = 100
gain = 0

# Appliquer la fonction √† chaque ligne du DataFrame fusionn√©
for index, row in merged_df.iterrows():
    gain += pari(row, mise)
    print('Le total des gains est de : ', gain)

mise -= 100
```

```python
# Afficher les r√©sultats
print(f"Nombre de fois o√π 'Prediction' est disponible avec les cotes : {non_nan_prediction_count}")
print(f"Nombre de fois o√π 'Prediction' n'est pas disponible avec les cotes: {nan_prediction_count}")

print(f"On se retrouve avec {gain}% de b√©n√©fice")
```
